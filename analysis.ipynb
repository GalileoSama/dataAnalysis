```python
import pandas as pd
import time
import numpy as np
```

```python
# coding=utf-8
# names=["数据key", "汽车id", "开始时间", "结束时间", "开始纬度", 
#                  "开始经度", "结束纬度", "结束经度"],
df = pd.read_csv('data/train.csv',  sep=",", low_memory=False)
df.head()
```

```python
# coding=utf-8
# 查看是否有重复数据  ——》无
df[df.duplicated()]
```

```python
# coding=utf-8
# 查看是否有Nah数据 ——》无
df[df.isnull().values == True]
```

```python
for i in list(df.columns):
    if df[i].dtype == object:
        print(i)
```

```python
df0 = df.copy()
```

```python
# coding=utf-8
# 时间转换为时间戳格式
df0['start_time'] = df['start_time'].apply(lambda x: time.mktime(time.strptime(x, '%Y-%m-%d %H:%M:%S'))).astype(int)

df0['end_time'] = df['end_time'].apply(lambda x: time.mktime(time.strptime(x, "%Y-%m-%d %H:%M:%S"))).astype(int)

```

```python
df1['drive_time'] = df0.end_time - df0.start_time
```
